Attempt to convert country names to standard country codes (like ISO 3166-1 alpha-3) for better map plotting.


**Reasoning**:
Import the necessary library and define a function to convert country names to ISO alpha-3 codes. Then, apply this function to the 'Country' column and display the updated DataFrame.



```python
import pycountry

def get_country_code(country_name):
    """
    Converts a country name to its ISO 3166-1 alpha-3 code.
    Handles multiple countries listed in a single string.
    """
    if pd.isna(country_name) or country_name.lower() == 'not listed':
        return 'Unknown'
    countries = [c.strip() for c in country_name.split(',')]
    codes = []
    for country in countries:
        try:
            # Try to find by common name
            country_data = pycountry.countries.lookup(country)
            codes.append(country_data.alpha_3)
        except LookupError:
            # If not found, try fuzzy matching
            try:
                fuzzy_country = pycountry.countries.search_fuzzy(country)
                if fuzzy_country:
                    codes.append(fuzzy_country[0].alpha_3)
                else:
                    codes.append('Unknown')
            except LookupError:
                codes.append('Unknown')

    # Return the code for the first country listed, or 'Unknown' if none found
    return codes[0] if codes and codes[0] != 'Unknown' else 'Unknown'

netflix_movies_sheet1_df_cleaned['Country_Code'] = netflix_movies_sheet1_df_cleaned['Country'].apply(get_country_code)

print("Netflix Movies Sheet 1 (Titles) with Country Codes Head:")
display(netflix_movies_sheet1_df_cleaned.head())
```

## Add tooltips/information

### Subtask:
Include tooltips or pop-ups on the map to show country names and the corresponding movie count when hovering over a country.


**Reasoning**:
The current `country_counts` is a Series, which is not ideal for directly using `hover_name` and `hover_data` in Plotly Express. To include the country names and movie counts in the tooltip, I will convert the `country_counts` Series into a DataFrame with explicit columns for country name and movie count. Then I will use this DataFrame to create the choropleth map with the specified hover parameters.



```python
# Convert the country_counts Series to a DataFrame for easier plotting with hover data
country_df = country_counts.reset_index()
country_df.columns = ['Country', 'Movie Count']

# Create the choropleth map with hover information
fig = px.choropleth(country_df,
                    locations='Country',
                    color='Movie Count',
                    locationmode='country names',
                    hover_name='Country',
                    hover_data=['Movie Count'],
                    title='Number of Netflix Movies per Country')

# Display the map
fig.show()
```
